```python
import warnings
warnings.filterwarnings("ignore", message="Wswiglal-redir-stdio")

from pycbc.waveform import get_td_waveform
import matplotlib.pyplot as plt
import numpy as np
import sxs
```

```python

print (sxs.__version__)
```

```python
m1 = 31.0
m2 = 20.5 

mtot = m1 + m2
sxsid = "SXS:BBH:0083"
hp, hc = get_td_waveform(approximant='nrsxs',
                                 mtotal=mtot,
                                 delta_t=1/512,
                                 f_lower=20,
                                 f_ref = 20,
                                 inclination=np.pi/6,
                                 coa_phase=np.pi/4,
                                 distance=1,
                                 sxs_id=sxsid)
plt.plot(hp.sample_times, hp)
plt.xlabel('Time (s)')
plt.ylabel('Strain') 
plt.grid()
plt.show()
```
